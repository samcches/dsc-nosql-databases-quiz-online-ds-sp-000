## Question 3

How does MongoDB work? How is it different from a relational database?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________


MongoDB is a document store. Although it is easier to store data in a document store than in a relational database, it is more difficult to query the data as each document can have its own schema. Each document stored in MongoDB or any other document store contain key-value pairs wherein the data is stored as the value. In a relational database, on the other hand, the schema is far more restrictive, with each document required to have the same keys as other documents.





## Question 4

What is fault tolerance, and what does it have to do with Resilient Distributed Datasets?

_______________________________________________________________________________________________________________________________



Fault tolerance is the RDD's way of making sure that all data from all servers are acessible at any given time. Making backups of each section of the database on each server on which the database is stored ensures that if the server(s) fails, the data is still available for analysis.
